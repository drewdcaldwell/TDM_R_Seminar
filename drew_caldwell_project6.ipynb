## Question 3

```r
library(data.table)
myDF <- fread("/anvil/projects/tdm/data/beer/reviews_sample.csv")

# Create a table of usernames and count
username_counts <- table(myDF$username)

# Sort table in descending order of count
sorted_counts <- sort(username_counts, decreasing = TRUE)

# Display only the tail
top6_usernames <- tail(sorted_counts, 6)
print(top6_usernames)
```

So these are the top 6 usernames

```r
# get the rows where the username is acurtis
acurtis_reviews <- myDF[myDF$username == "acurtis", ]

# find avg score for acurtis
average_score_acurtis <- mean(acurtis_reviews$score, na.rm = TRUE)

# display the average score
cat("The average score for acurtis is:", average_score_acurtis, "\n")
```

The average score for the username acurtis is 3.65

## Question 4

```r
library(data.table)
myDF <- fread("/anvil/projects/tdm/data/election/itcont1980.txt", quote="")

# designate column names
names(myDF) <- c("CMTE_ID", "AMNDT_IND", "RPT_TP", "TRANSACTION_PGI", "IMAGE_NUM", 
                 "TRANSACTION_TP", "ENTITY_TP", "NAME", "CITY", "STATE", "ZIP_CODE", 
                 "EMPLOYER", "OCCUPATION", "TRANSACTION_DT", "TRANSACTION_AMT", 
                 "OTHER_ID", "TRAN_ID", "FILE_NUM", "MEMO_CD", "MEMO_TEXT", "SUB_ID")

# filter the rows where TRANSACTION_DT is missing 
missing_dates <- myDF[is.na(myDF$TRANSACTION_DT), ]

# get the NAME values for those rows
missing_names <- missing_dates$NAME

#print the first 9 different NAME values
unique_missing_names <- unique(missing_names)[1:9]
print(unique_missing_names)
```

## Question 5

```r
library(data.table)
myDF <- fread("/anvil/projects/tdm/data/flights/subset/1990.csv")

# filter the flights that start at EWR, JFK, LGA
filtered_flights <- myDF[myDF$Origin %in% c("EWR", "JFK", "LGA"), ]

# get the mean of DepDelay for the flights filtered from above
mean_dep_delay <- mean(filtered_flights$DepDelay, na.rm = TRUE)

# dsiplay the mean DepDelay
cat("The mean DepDelay for flights originating from EWR, JFK, or LGA is:", mean_dep_delay, "\n")
```
